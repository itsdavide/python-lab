```python
# Import libraries
import numpy as np
import pandas as pd
```

```python
# Load the dataset
return_rates = pd.read_csv('GARR.csv')
return_rates
```

```python
# Show the plot
return_rates.plot()
```

```python
# Show the histogram
return_rates.hist()
```

```python
# Compute the linear regression
from scipy import stats

linear_regression = stats.linregress(x = return_rates['AMZN'], 
                                     y = return_rates['GOOG'])
a = linear_regression.slope
b = linear_regression.intercept
```

```python
# Plot the regression line using Seaborn library
import seaborn as sn

axes = sn.regplot(x = return_rates['AMZN'], y = return_rates['GOOG'])
```

```python
# Define a function to compute a prediction with the linear regression
def prediction(x, a, b):
    return x * a + b
```

```python
# Evaluate the prediction with linear regression
prediction(0, a, b)
```
