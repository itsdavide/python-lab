```python
import random

# Variables for the absolute frequency of numbers
frequency1 = 0
frequency2 = 0
frequency3 = 0
frequency4 = 0
frequency5 = 0
frequency6 = 0

# Number of samples
N = 6000000
```

```python
# N die rolls
for roll in range(N):
    # Roll the dice
    face = random.randrange(1, 7)
    
    if face == 1:
        frequency1 += 1
    elif face == 2:
        frequency2 += 1
    elif face == 3:
        frequency3 += 1
    elif face == 4:
        frequency4 += 1
    elif face == 5:
        frequency5 += 1
    elif face == 6:
        frequency6 += 1
```

```python
print('Face 1 =', frequency1 / N)
print('Face 2 =', frequency2 / N)
print('Face 3 =', frequency3 / N)
print('Face 4 =', frequency4 / N)
print('Face 5 =', frequency5 / N)
print('Face 6 =', frequency6 / N)
```

```python
# Create a list of relative frequencies
rfreq = [frequency1 / N, frequency2 / N, frequency3 / N, frequency4 / N, frequency5 / N, frequency6 / N]
rfreq
```

```python
# Create a list of labels
labels = ['Face 1', 'Face 2', 'Face 3', 'Face 4', 'Face 5', 'Face 6']
labels
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a figure
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_ylabel('Relative frequency')
ax.set_title('Relative frequency with ' + str(N) + ' rolls')

ax.bar(labels, rfreq)
plt.show()
```
